```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
from google.colab import files

# Upload the CSV file
uploaded = files.upload()

# Load the uploaded CSV file (the first file uploaded)
for filename in uploaded.keys():
    df = pd.read_csv(filename)

# Select the relevant features for clustering: age and potential
X = df[['age', 'potential']].values

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)
labels = kmeans.labels_

# Add the cluster labels to the DataFrame
df['Cluster'] = labels

# Plot the clusters with colorbar
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df['age'], df['potential'], c=df['Cluster'], cmap='viridis')
plt.xlabel('Age')
plt.ylabel('Potential')
plt.title('KMeans Clustering of FIFA Players by Age and Potential')

# Add color bar to indicate cluster groups
plt.colorbar(scatter)
plt.show()
```
